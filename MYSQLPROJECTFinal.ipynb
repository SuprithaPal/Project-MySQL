## Database Connectivity:

To perform the analysis, we'll establish a connection to the MySQL database containing
the ecommerce data. We'll use the appropriate Python libraries, such as pandas and
mysql, to fetch and manipulate the data directly from the database

```python
import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def mydbconnect():
     mydb = mysql.connector.connect (
     host="localhost",
     user="root",
     password="root",
     database="e_commerce"
      )
     return mydb
```

```python
mydb=mydbconnect()
mycursor = mydb.cursor()
# Query data from the 'customer' table
mycursor.execute("SELECT * FROM customer")
#After fetching data from the database we are storing it into Pandas DataFrame
customer_data = pd.DataFrame(mycursor.fetchall(), columns=[desc[0] for desc in mycursor.description])
#printing first 5 records 
print("First 5 records from the 'customer' table:")
print(customer_data.head())


mydb.close()
```

```python
mydb=mydbconnect()
mycursor = mydb.cursor()
# Query data from the 'products' table
mycursor.execute("SELECT * FROM products")
products_data = pd.DataFrame(mycursor.fetchall(),columns=[desc[0] for desc in mycursor.description])
#printing first 5 records 
print("First 5 records from the 'Products' table:")
print(products_data.head())


mydb.close()
```

```python
mydb=mydbconnect()
mycursor = mydb.cursor()
# Query data from the 'orderdetails' table
mycursor.execute("SELECT * FROM orderdetails")
orderdetails_data = pd.DataFrame (mycursor.fetchall(),columns=[desc[0] for desc in mycursor.description])
#printing first 5 records 
print("First 5 records from the 'orderdetails' table:")
print(orderdetails_data.head())
  

mydb.close()
```

## Data Cleaning:


 Before proceeding with the analysis, let's perform some basic data cleaning:

```python
#Removing missing values:
print("Database info from customer table:")
print(customer_data.info())      
print("Database info from products table:")
print(products_data.info()) 
print("Database info from orderdetails table:")
print(orderdetails_data.info()) 
```

```python
#Check for missing values :
print("Mising values in customer table:")
print(customer_data.isnull().sum())
print("Mising values in products table:")
print(products_data.isnull().sum())
print("Mising values in orderdetails table:")
print(orderdetails_data.isnull().sum())
```

```python
#Drop columns with any missing values:
print(customer_data.dropna(inplace=True))
print(products_data.dropna(inplace=True))
print(orderdetails_data.dropna(inplace=True))
```

```python
#Data after cleaning:
print("After removing missing value from customer table:")
print(customer_data.info())
print("After removing missing value from products table:")
print(products_data.info())
print("After removing missing value from orderdetails table:")
print(orderdetails_data.info())
```

```python
#Remove duplicate valus:
print("After removing duplicate value from customer table:")
print(customer_data.drop_duplicates(inplace=True))
print("After removing duplicate value from products table:")
print(products_data.drop_duplicates(inplace=True))
print("After removing duplicate value from orderdetails table:")
print(orderdetails_data.drop_duplicates(inplace=True))
```

## Exploratory Data Analysis (EDA) and Visualization:

### ‚óè Customer Analysis:


```python
# Identify the total number of customers City wise.
colors=['blue','seagreen','red','yellow']
City_wise_total_number_of_customers= customer_data.groupby('city')['customer_id'].count().sort_values(ascending=False)
City_wise_total_number_of_customers.plot(kind='bar',color=colors)
plt.title('City wise total number of customers')
plt.ylabel('Number of Customers')
plt.xlabel('City')
plt.xticks(rotation=45,ha='right')


plt.show()
```

```python
# Identify the most frequent customers based on their order history:
colors=['pink','yellow','skyblue','brown','green']
Most_frequent_customers_based_on_their_order_history=orderdetails_data.groupby('customer_id')['order_id'].count().sort_values(ascending=False)
Most_frequent_customers_based_on_their_order_history.plot(kind='bar',color=colors)
plt.title('Top 5 most frequent customers ')
plt.ylabel('Number of Orders')
plt.xlabel('Customer id')


plt.show()
```

### ‚óè Product Analysis:

```python
#Determine the total number of products available by category:
colors=['purple','yellow','green']
category_wise_total_number_of_products_available=products_data.groupby('category')['product_id'].count().sort_values(ascending=False)
category_wise_total_number_of_products_available.plot(kind='bar',color=colors)
plt.title('Distribution of products')
plt.ylabel('Number of occurences')
plt.xlabel('Product id')


plt.show()
```

```python
#Analyze the distribution of products across sub-categories:
colors=['maroon','skyblue','red','yellow','green']
distribution_of_products_across_sub_categories=products_data.groupby('sub_category')['product_id'].count().sort_values(ascending=False)
distribution_of_products_across_sub_categories.plot(kind='bar',color=colors)
plt.title('Distribution of products across sub-categories')
plt.ylabel('Number of products')
plt.xlabel('Sub Category')
plt.xticks(rotation=45,ha='right')


plt.show()
```

```python
#Identify products with low stock levels:
colors=['red','blue','coral']
Identify_products_with_low_stock_levels=products_data.groupby('product_name')['stock'].sum().sort_values(ascending=False)
print(Identify_products_with_low_stock_levels)
Identify_products_with_low_stock_levels.plot(kind='bar',color=colors)
plt.title('Identify products with low stock levels')
plt.xlabel('Stock level')
plt.ylabel('Product name')
plt.xticks(rotation=45,ha='right')

plt.show()
```

```python
#Calculate the average, maximum, and minimum selling prices for products:
average_selling_price = products_data['selling_price'].mean()
max_selling_price = products_data['selling_price'].max()
min_selling_price = products_data['selling_price'].min()
print('average_selling_price',average_selling_price)
print('max_selling_price',max_selling_price)
print('min_selling_price',min_selling_price)
colors=['skyblue', 'orange', 'green']
plt.bar(['Average', 'Maximum', 'Minimum'], [average_selling_price, max_selling_price, min_selling_price], color=colors)
plt.ylabel('Price')
plt.title('Average, Maximum, and Minimum Selling Price for Products')
plt.tight_layout()
plt.show()

```

### ‚óè Order Analysis:

```python
#Calculate the top 6 orders product wise:
mydb=mydbconnect()
sqlstmt="select o.order_id, o.quantity,p.product_name from products p natural join orderdetails o"
sql_query = pd.read_sql_query(sqlstmt,mydb)
df = pd.DataFrame(sql_query, columns = ['order_id','quantity','product_name'])
colors=['red','blue','pink','grey','coral']
df.groupby('product_name')['order_id'].sum().sort_values(ascending=False).plot(kind='bar',color=colors)
plt.title('Total number of orders Products with Product-names')
plt.ylabel('Number of orders')
plt.xlabel('Product name')
plt.xticks(rotation=45,ha='right')

plt.show() 
mydb.close()
```

```python
# Analyse the order status distribution:
colors=['purple','violet','pink']
order_status_distribution=orderdetails_data.groupby('order_status')['order_id'].count().sort_values(ascending=False)
order_status_distribution.plot(kind='bar',color=colors)
plt.title('Order Status Distribution')
plt.ylabel('Number of orders')
plt.xlabel('Order Status')


plt.show() 
```

```python
#Identify the most popular products based on order quantity:
mydb=mydbconnect()
sqlstmt="select o.order_id, o.quantity,p.product_name from products p natural join orderdetails o"
sql_query = pd.read_sql_query(sqlstmt,mydb)
df = pd.DataFrame(sql_query, columns = ['order_id','quantity','product_name'])
colors=['red','blue','pink','seagreen','coral']
df.groupby('product_name')['quantity'].sum().sort_values(ascending=False).plot(kind='barh',color=colors)
plt.title('Top most popular products based on order quantity')
plt.ylabel('Total Order quantity')
plt.xlabel('Product name')


plt.show() 
mydb.close()
```

### ‚óè Sales Analysis:

```python
#Calculate total revenue generated from orders product wise:
mydb=mydbconnect()
sqlstmt="select o.order_id, o.quantity,o.total_price,p.product_name from products p natural join orderdetails o"
sql_query = pd.read_sql_query(sqlstmt,mydb)
df = pd.DataFrame(sql_query, columns = ['order_id','quantity','product_name','total_price'])
colors=['brown','yellow','orange','green','purple']
df.groupby('product_name')['total_price'].sum().sort_values(ascending=False).plot(kind='bar',color=colors)
plt.title('Top 5 products by total revenue')
plt.xlabel('Product name')
plt.ylabel('Total revenue')
plt.xticks(rotation=45,ha='right')
plt.show() 
mydb.close()
```

```python
#Calculate total revenue product category wise percentage:
mydb=mydbconnect()
sqlstmt="select o.order_id, o.quantity,o.total_price,p.product_name,p.category from products p natural join orderdetails o"
sql_query = pd.read_sql_query(sqlstmt,mydb)
df = pd.DataFrame(sql_query, columns = ['order_id','quantity','product_name','total_price','category'])
colors=['brown','yellow','orange','green','purple']
df.groupby('category')['total_price'].sum().sort_values(ascending=False).plot(kind='pie',autopct='%1.2f%%',)
plt.title('Total revenue percentage for each category')
plt.show() 
mydb.close()
```

```python
#Calculate the total revenue generated from all orders:
Total_revenue=orderdetails_data['total_price'].sum()
plt.bar('Total Revenue Metrics',Total_revenue)
plt.ylabel("Total Revenue")
plt.title("Total Revenue Generated from all Orders")
plt.show()
```

```python
#Identify the most profitable products based on the difference between original and selling prices:
mydb=mydbconnect()
sqlstmt="select o.order_id, o.quantity,o.total_price,p.product_name,p.category,p.original_price ,p.selling_price - p.original_price AS profit from products p natural join orderdetails o"
sql_query = pd.read_sql_query(sqlstmt,mydb)
df = pd.DataFrame(sql_query, columns = ['order_id','quantity','product_name','total_price','category','profit'])
colors=['brown','yellow','orange','green','purple']
df.groupby('product_name')['profit'].sum().sort_values(ascending=False).plot(kind='bar',color=colors)
plt.title('Top 4 most profitable products')
plt.xlabel('Product name')
plt.ylabel('Profit(selling price - original price)')
plt.xticks(rotation=45,ha='right')
plt.show() 
mydb.close()
```

### ‚óè Customer Order Patterns:

```python
# Identify product names with the highest and lowest order quantities:
mydb=mydbconnect()
mycursor = mydb.cursor()
highest_query = "SELECT p.product_name, SUM(o.quantity) AS total_quantity FROM orderdetails o JOIN products p ON o.product_id = p.product_id GROUP BY p.product_name ORDER BY total_quantity DESC limit 1"
lowest_query = "SELECT p.product_name, SUM(o.quantity) AS total_quantity FROM orderdetails o JOIN products p ON o.product_id = p.product_id GROUP BY p.product_name ORDER BY total_quantity ASC limit 1"

mycursor.execute(highest_query)
highest_quantity_product = mycursor.fetchone()

mycursor.execute(lowest_query)
lowest_quantity_product = mycursor.fetchone()

df = pd.DataFrame([highest_quantity_product,lowest_quantity_product],columns = ['product_name','total_quantity'])
colors=['blue','red']
plt.bar(df['product_name'], df['total_quantity'],color=colors)
plt.ylabel('Total Quantity Ordered')
plt.xlabel('Product Name')
plt.title('Total quantity of Highest and Lowest Ordered Products')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
#Identify customers with the highest and lowest order quantities by customer name.
mydb=mydbconnect()
mycursor = mydb.cursor()
highest_query = "SELECT c.name, SUM(o.quantity) AS total_quantity FROM orderdetails o JOIN customer c ON o.customer_id = c.customer_id GROUP BY c.name ORDER BY total_quantity DESC limit 1"
lowest_query = "SELECT c.name, SUM(o.quantity) AS total_quantity FROM orderdetails o JOIN customer c ON o.customer_id = c.customer_id GROUP BY c.name ORDER BY total_quantity ASC limit 1"

mycursor.execute(highest_query)
highest_quantity_product = mycursor.fetchone()

mycursor.execute(lowest_query)
lowest_quantity_product = mycursor.fetchone()

df = pd.DataFrame([highest_quantity_product,lowest_quantity_product],columns = ['customer_name','total_quantity'])
colors=['seagreen','violet']
plt.bar(df['customer_name'], df['total_quantity'],color=colors)
plt.ylabel('Total Ordered Quantity')
plt.xlabel('Customer Name')
plt.title('Total order quantity by customer ( Highest and Lowest)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### ‚óè Time-based Analysis:

```python
# Month wise total sales:
orderdetails_data['order_date'] = pd.to_datetime(orderdetails_data['order_date'])
orderdetails_data['month'] = orderdetails_data['order_date'].dt.strftime('%B')
colors=['violet','coral']
orderdetails_data.groupby('month')['total_price'].sum().sort_values(ascending=False).plot(kind='bar',color=colors)
plt.title('Month wise total sales')
plt.xlabel('Month')
plt.ylabel('Total sales')
plt.xticks(rotation=45,ha='right')
plt.show() 
mydb.close()
```

```python
#Month and year wise total sales
orderdetails_data['order_date'] = pd.to_datetime(orderdetails_data['order_date'])
orderdetails_data['month_year'] = orderdetails_data['order_date'].dt.strftime('%B %Y')
colors=['seagreen','coral']
orderdetails_data.groupby('month_year')['total_price'].sum().sort_values(ascending=False).plot(kind='bar',color=colors)
plt.title('Year wise and Month wise total sales')
plt.xlabel('Month and Year')
plt.ylabel('Total sales')
plt.xticks(rotation=45,ha='right')
plt.show() 
mydb.close()
```

```python
#Identify peak order date.
orderdetails_data.groupby('order_date')['quantity'].sum().sort_values(ascending=False).plot(kind='line',marker='o')
plt.title('Peak Order Periods')
plt.xlabel('Order date')
plt.ylabel('Number of orders')
plt.xticks(rotation=45,ha='right')
plt.show() 
mydb.close()
```

### ‚óè Product Performance:
